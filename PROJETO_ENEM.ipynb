## Referencial teórico: Teoria Geral dos Campos de Pierre Bourdieu

Ao olharmos para um conjunto de dados, não podemos fazê-lo às cegas. É necessária uma teoria, uma lente, para direcionar a pesquisa. ### Campo

É um espaço social específico onde indivíduos e instituições interagem e competem por recursos e poder. Cada campo possui suas próprias regras, normas e formas de avaliação que determinam como os "jogadores" devem agir para serem reconhecidos ou alcançarem posições de destaque. Exemplos de campos incluem o campo artístico, o campo científico, o campo político e o campo educacional.

### Habitus

É um sistema de disposições duráveis e internalizadas que moldam percepções, ações e pensamentos dos indivíduos. Ele é formado pela experiência social e pela posição do indivíduo no campo social, sendo influenciado pelas condições históricas e culturais em que a pessoa foi socializada.

### Capital

O capital, em Bourdieu, vai além do econômico. Ele é o recurso que permite aos agentes se posicionarem e competirem nos campos sociais. Bourdieu identifica quatro principais tipos de capital:

Capital Econômico: Recursos materiais e financeiros, como renda e patrimônio.

Capital Cultural: Conhecimentos, habilidades e credenciais educacionais que permitem o acesso a certos campos ou posições sociais. Por exemplo, um diploma de prestígio.

Capital Social: Rede de relações e conexões que podem ser mobilizadas para alcançar objetivos.

Capital Simbólico: Reconhecimento, prestígio ou legitimidade social que valida as outras formas de capital.

## Hipóteses

Em estudo na França, Bourdieu demonstra fatores que podemos testar em nossa realidade:

i) Dentre todos os capitais, o mais relevante é o financeiro: é aquele capital que, caso o indivíduo possua, pode ter acesso aos outros capitais. Portanto supõe-se que seja o maior indicador das notas

ii) a escolaridade dos estudantes costuma reproduzir a escolaridade dos pais: assim sendo, a educação acaba reproduzindo as desigualdade sociais (alunos que fazem faculdade geralmente têm pais que fizeram faculdade e assim por diante)

**Em suma: seria de fato o capital financeiro o principal fatov preditivo das notas no ENEM?**

### Especificidades do problema

Aqui, buscando saber se a educação de fato, em nossa base de dados, costuma reproduzir as desigualdades sociais, dividirei as notas dos alunos em patamares. Por exemplo: se a nota para passar em um determinado curso é de 750, qualquer nota igual ou acima disso o classifica. Nesse sentido, não se faz necessário saber a nota exata do estudante, mas sim se ele ou ela atingiriam a nota mínima para passar em um curso.


Buscando patamares realistas, mas não sendo exaustivo na avalaiação de notas de corte, os alunos serão divididos nos seguintes grupos:

- Grupo 1: Alunos que conseguiriam vaga na maior parte dos cursos mais concorridos do SISU. **Nota de Corte: 720**
- Grupo 2: Alunos que conseguiriam vaga na maior parte dos cursos fora dos 10 mais concorridos do SISU. **Nota de Corte: 650**
- Grupo 3: Alunos que conseguiriam vaga apenas em alguns cursos do SISU, nos menos concorridos. **Nota de Corte: 500**
- Grupo 4: Alunos que não conseguiriam vaga em ampla concorrência na maior parte das vagas. **Nota de Corte: < 500**


* Nota final: é calculada a partir da média simples das 5 provas -> Matemática, Linguagens e códigos, Ciências da natureza, Ciências humanas e Redação.

```python
def print_metricas(dados, 
                  PD = 'PD', 
                  CLASSE_PRED = 'classe_predita', 
                  RESP = 'mau'):
    
    #Acuracia
    acc = accuracy_score(dados[RESP], dados[CLASSE_PRED])

    #AUC
    auc = roc_auc_score(dados[RESP], dados[PD])
    
    #Gini
    gini = 2*auc-1
    
    #KS
    ks = ks_2samp(dados.loc[dados[RESP] == 1, PD], 
                  dados.loc[dados[RESP] != 1, PD]).statistic

    print('KS:       {0:.2f}%'.format(ks*100))
    print('AUC:      {0:.2f}%'.format(auc*100))
    print('GINI:     {0:.2f}%'.format(gini*100))
    print('Acurácia: {0:.2f}%\n'.format(acc*100))
    
    return None
```

### Gridsearch

```python
%%time
from sklearn.preprocessing import LabelEncoder 

le = LabelEncoder()
```

```python
%%time
import xgboost as xgb
X_train1 = pd.get_dummies(X_train1, drop_first=True)
X_test1 = df_test1.drop(columns = 'grupos_sisu')

y_test1 = df_test1['grupos_sisu']
le = LabelEncoder()
y_train1_encoded = le.fit_transform(y_train1)
y_test1_encoded = le.transform(y_test1)  # Use transform, não fit_transform

X_test1 = pd.get_dummies(X_test1, drop_first=True)
# Atualizar as variáveis
dtrain = xgb.DMatrix(X_train1, label=y_train1_encoded,enable_categorical= True)
dtest = xgb.DMatrix(X_test1, label=y_test1_encoded,enable_categorical= True)
```

```python
params = {
    'objective': 'multi:softprob',  # Para classificação multiclasse
    'num_class': len(y_train1.unique()),   # Número de classes (subgrupos)
    'learning_rate': 0.1,           # Taxa de aprendizado (eta)
    'max_depth': 6,                 # Profundidade máxima das árvores
    'subsample': 0.8,               # Amostragem de linhas
    'colsample_bytree': 0.8,        # Amostragem de colunas
    'eval_metric': 'mlogloss',      # Métrica de avaliação
    'seed': 42                      # Reprodutibilidade
}
```

```python
# Treinar com validação para evitar overfitting
evals = [(dtrain, 'train'), (dtest, 'eval')]
model = xgb.train(
    params,
    dtrain,
    num_boost_round=1000,  # Número máximo de árvores
    evals=evals,
    early_stopping_rounds=20,  # Parar se não houver melhora em 20 iterações
    verbose_eval=10
)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Prever no conjunto de teste
y_pred_prob = model.predict(dtrain)
y_pred = y_pred_prob.argmax(axis=1)  

y_pred
```

```python
y_pred_labels = le.inverse_transform(y_pred)

# Métricas de avaliação
print(f"Acurácia: {accuracy_score(y_train1, y_pred_labels):.2f}")
print(classification_report(y_train1, y_pred_labels, target_names=le.classes_))
```
